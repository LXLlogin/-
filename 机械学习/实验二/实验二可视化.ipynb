```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# 加载鸢尾花卉数据集 filename(数据集文件存放路径)
data = []
filename = "C:\\Users\\LXL\\Desktop\\data实验2-课程1成绩，体能成绩.txt"
with open(filename, 'rt') as csvfile:
    lines = csv.reader(csvfile)
    data = list(lines)

list_x=data[0]
list_x.pop()
list_y=data[1]
list_y.pop()

ax = plt.subplot()
ax.set_xlabel('course 1 score')
ax.set_ylabel('sports score')
ax.set_title('course 1 score--sports score')
plt.scatter(list_x,list_y,c='blue',marker='*')
plt.show()

list_x.sort()
ax = plt.subplot()
ax.set_xlabel('course 1 score')
plt.hist(list_x, bins=5, edgecolor='k', alpha=0.35) # 设置直方边线颜色为黑色，不透明度为 0.35
plt.show()
```
